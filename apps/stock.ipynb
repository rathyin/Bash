```python
from jupyter_dash import JupyterDash
import dash
import dash_bootstrap_components as dbc
import dash_html_components as html
import dash_core_components as dcc
import plotly.express as px
from dash.dependencies import Input, Output
import datetime as dt
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
from datetime import date as dt, timedelta
from ipynb.fs.full.app  import app

layout = html.Div([
     html.H1('Stock Analysis', style={"textAlign": "center"}), 
    html.Div([
        
        html.Div([dcc.DatePickerRange(
        id='my-date-picker-range',
        display_format='DD-MM-YYYY',
        first_day_of_week=1,
        max_date_allowed=dt.today(),
    )],className = "drop1") ,  
        html.Div([dcc.Dropdown(
        id='dropdown',
        options=[
            {'label': 'Google', 'value': 'GOOG'},
            {'label': 'Microsoft', 'value': 'MSFT'},
            {'label': 'Apple', 'value': 'AAPL'},
            {'label': 'Amazon', 'value': 'AMZN'},
            {'label': 'Facebook', 'value': 'FB'},
            {'label': 'Tesla', 'value': 'TSLA'},
        ],
        value='GOOG'
    )],className = "drop2"),
        html.Div([dcc.Dropdown(
        id='dropdownchart',
        options=[
            {'label': 'LineChart', 'value': 'LineChart'},
            {'label': 'Candlestick', 'value': 'Candlestick'},
        ],
        value='LineChart'
    )],className = "drop3"),
    ],className = "outerdiv"),

# html.Div([dcc.Graph(id='mybar', figure={})],className = "chart")
    dcc.Graph(id='mybar', figure={}),
],className = "graph")

# layout = html.Div([
#     dcc.DatePickerRange(
#         id='my-date-picker-range',
#         display_format='DD-MM-YYYY',
#         first_day_of_week=1,
#         max_date_allowed=dt.today(),
#     ),
#     html.Div(id='page-content', children=[])
# ])

@app.callback(
    dash.dependencies.Output('mybar', 'figure'),
    [dash.dependencies.Input('my-date-picker-range', 'start_date'),
     dash.dependencies.Input('my-date-picker-range', 'end_date'),
    dash.dependencies.Input('dropdown', 'value'),
    dash.dependencies.Input('dropdownchart', 'value'),])
def updateDataPicker(start_date,end_date,company,chart):
    df = yf.download(company,start = start_date,end = end_date)
    if(chart == "LineChart"):
        df=df.transpose()
        df.rename(index={"1. open":"open", "2. high":"high", "3. low":"low",
                     "4. close":"close","5. Adj_Close":"Adj Close","6. Volume":"Volume"},inplace=True)
        df=df.reset_index().rename(columns={'index': 'indicator'})

        df = pd.melt(df,id_vars=['indicator'],var_name='date',value_name='rate')
        df = df[df['indicator']!='Volume']
        fig = px.line(
                        data_frame=df,
                        x='date',
                        y='rate',
                        color='indicator',
                        title=" "
                     )
        return fig
    if(chart == "Candlestick"):
        df.rename(index={"1. open":"open", "2. high":"high", "3. low":"low",
               "4. close":"close","5. Adj_Close":"Adj Close","6. Volume":"Volume"},inplace=True)
        df=df.reset_index().rename(columns={'index': 'indicator'})
        fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
        return fig
    else:
        return 0
    
```
